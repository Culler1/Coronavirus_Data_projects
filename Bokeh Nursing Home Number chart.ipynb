```python
import pandas as pd
import datetime as DT
from datetime import timedelta
import numpy as np
from bokeh.io import output_notebook
from bokeh.plotting import figure, show
from bokeh.plotting import reset_output
from bokeh.models import ColumnDataSource, CDSView, GroupFilter
from bokeh.models import HoverTool,ColumnDataSource,LinearColorMapper,GeoJSONDataSource, ColorBar,LogTicker,BasicTicker
output_notebook()
reset_output()
```

```python
df=pd.read_csv("df1.csv")
df['date']=pd.to_datetime(df.date)
bucksdata2=df[df['county']=='Bucks']
bucks_cds3=ColumnDataSource(bucksdata2)
#southeasternpa=['Bucks','Berks','Montgomery','Chester','Delaware','Lancaster','Philadelphia','Schuylkill']
```

```python
output_notebook()
fig = figure(x_axis_type='datetime',background_fill_color='silver',
             plot_height=800, plot_width=950,
             title='Bucks County coronavirus cases per 100,000 people over the past two weeks',
             x_axis_label='Date', y_axis_label='current_rate',
             toolbar_location=None)
fig.line('date', 'current_rate', legend_label='Bucks Total Cases per 100,000 people',
         color='red', line_width=2,
         source=bucks_cds3)
fig.dash('date', 'possible_rate', legend_label='Bucks Minus Nursing Home Residents Cases per 100,000 people',
         color='blue', line_width=2,line_dash="solid",size=20,
         source=bucks_cds3)
fig.line('date', 'rate_goal', legend_label='Bucks Minus Nursing Home Residents Cases per 100,000 people',
         color='green', line_width=3,
         source=bucks_cds3)
fig.legend.location='top_left'
tooltips=[('Current Rate',"@current_rate{0.00}"),('Possible Rate','@possible_rate'),("Cases","@cases"),('Sick Residents',"@sick_residents"),
          ('Population',"@population{0,000.00}"),('Statwide Benchmark','@rate_goal'),('Date',"@date{%F}")]
my_hover=HoverTool
fig.add_tools(HoverTool(tooltips=tooltips,formatters={'@date': 'datetime'}))
show(fig)
```
